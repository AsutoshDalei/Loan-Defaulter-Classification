#Data Cleaning

Checking for % null records in all columns

```python
null_columns=df.columns[df.isnull().any()]
missing_value_columns=[]
for i in null_columns:
    print('% of Nulls in dataset for column',i,'is ',round(df[i].isnull().sum()/df[i].shape[0]*100,2))
    null_percent=round(df[i].isnull().sum()/df[i].shape[0]*100,2)
    missing_value_columns.append([i,null_percent])

#there a lot of columns with more than 50% of the rows empty
```

Getting the columns that have more than 50% missing data

```python
fifty_percent_missing_columns=[]
for i in missing_value_columns:
  if i[1]>=50:
    fifty_percent_missing_columns.append(i[0])
print(fifty_percent_missing_columns)
```

It does not make sense to fill missing values in columns that have more than 50% of their data missing. So we'll drop these columns

```python
for i in fifty_percent_missing_columns:
    df.drop(i,axis=1,inplace=True)
```

Finding out how many columns we actually dropped

```python
print(len(fifty_percent_missing_columns))
```

Number of columns dropped that had more than 50% missing data - 41

Dropping the SKU ID column as well

```python
df=df.drop('SK_ID_CURR',axis=1)
```

Focusing on the columns with less than 50% missing data

```python
lessthan_50percent_missing = [col[0] for col in missing_value_columns if col[0] not in fifty_percent_missing_columns]
print(lessthan_50percent_missing)
```

```python
lessthan_5percent_missing = [col[0] for col in missing_value_columns if col[1]<5]
print(lessthan_5percent_missing)
```

The above outputs shows the highly skewed distribution for DEF_30_CNT_SOCIAL_CIRCLE and DEF_60_CNT_SOCIAL_CIRCLE. We'll remove outliers using the IQR logic and check again

```python
for column in ['OBS_30_CNT_SOCIAL_CIRCLE','OBS_60_CNT_SOCIAL_CIRCLE','DEF_30_CNT_SOCIAL_CIRCLE','DEF_60_CNT_SOCIAL_CIRCLE']:
        # Calculate Q1 (25th percentile) and Q3 (75th percentile)
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1

        # Define lower and upper bounds
        lower_bound = Q1 - 1.5 * IQR
        if lower_bound<0:
          lower_bound=0
        upper_bound = Q3 + 1.5 * IQR
        print('new lower and upper bound for',i,' is',lower_bound,upper_bound)
        # Clip the outliers
        df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)

```
